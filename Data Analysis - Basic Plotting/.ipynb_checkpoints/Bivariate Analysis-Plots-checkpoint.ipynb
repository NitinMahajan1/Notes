## Heat Map
A heat map is a 2-d version of the histogram that can be used as an alternative to a scatterplot. Like a scatterplot, the values of the two numeric variables to be plotted are placed on the plot axes. Similar to a histogram, the plotting area is divided into a grid and the number of points in each grid rectangle is added up. Since there won't be room for bar heights, counts are indicated instead by grid cell color. A heat map can be implemented with Matplotlib's hist2d function.

```python
plt.figure(figsize=[20,5])

# Plot on Left: Scatter Plot of discrete data with jitter and transparency
plt.subplot(1,3,1)
sns.regplot(data=fuel_econ,x = 'displ', y='comb', fit_reg=False, x_jitter=0.2, y_jitter=0.2, scatter_kws={'alpha':1/3})

# Plot at Center: Heat Map with bin edges between values
plt.subplot(1,3,2)
bins_x = np.arange(0,fuel_econ.displ.max()+1,1)
bins_y = np.arange(0,fuel_econ.comb.max()+1,1)
plt.hist2d(data=fuel_econ, x ='displ', y='comb', bins =[bins_x,bins_y])
plt.colorbar()
 

# Plot on Right: Heat Map with bin edges between values
plt.subplot(1,3,3)
bins_x = np.arange(0,fuel_econ.displ.max()+1,1)
bins_y = np.arange(0,fuel_econ.comb.max()+1,1)
#an example of reversing the default "viridis" color palette, by setting cmap = 'viridis_r'
#distinguish cells with zero counts from those with non-zero counts
#By adding a cmin = 0.5 parameter to the hist2d call, this means that a cell will only get colored if it contains at least one point.
plt.hist2d(data=fuel_econ, x ='displ', y='comb',cmap='viridis_r', cmin=0.5)
plt.colorbar()    
```

```python
#plotting object, but an array of counts and two vectors of bin edges.
# hist2d returns a number of different variables, including an array of counts

plt.figure(figsize=[20,5])
plt.subplot(1,2,1)

bins_x = np.arange(0,fuel_econ.displ.max()+1,1)
bins_y = np.arange(0,fuel_econ.comb.max()+10,10)
h2d = plt.hist2d(data=fuel_econ, x ='displ', y='comb', bins =[bins_x,bins_y],cmap='viridis_r', cmin=0.5)
plt.colorbar()

counts = h2d[0]

# loop through the cell counts and add text annotations for each
for i in range(counts.shape[0]):
    for j in range(counts.shape[1]):
        c = counts[i,j]
        if c >= 7: # increase visibility on darkest cells
            plt.text(bins_x[i]+0.5, bins_y[j]+5, int(c),
                     ha = 'center', va = 'center', color = 'white')
        elif c > 0:
            plt.text(bins_x[i]+0.5, bins_y[j]+5, int(c),
                     ha = 'center', va = 'center', color = 'black')

#plotting object, but an array of counts and two vectors of bin edges.
# hist2d returns a number of different variables, including an array of counts
plt.subplot(1,2,2)
bins_x = np.arange(0,fuel_econ.displ.max()+0.5,0.5)
bins_y = np.arange(0,fuel_econ.comb.max()+5,5)
h2d = plt.hist2d(data=fuel_econ, x ='displ', y='comb', bins =[bins_x,bins_y],cmap='viridis_r', cmin=0.5)

counts = h2d[0]

# loop through the cell counts and add text annotations for each
for i in range(counts.shape[0]):
    for j in range(counts.shape[1]):
        c = counts[i,j]
        if c >= 7: # increase visibility on darkest cells
            plt.text(bins_x[i]+0.25, bins_y[j]+2.5, int(c),
                     ha = 'center', va = 'center', color = 'white')
        elif c > 0:
            plt.text(bins_x[i]+0.25, bins_y[j]+2.5, int(c),
                     ha = 'center', va = 'center', color = 'black')
```

```python
fuel_econ['trans_type'] = fuel_econ['trans'].apply(lambda x : x.split()[0])
```

```python
ct_counts = fuel_econ.groupby(['VClass','trans_type']).size()
ct_counts = ct_counts.reset_index(name='count')
ct_counts
```

```python
ct_counts  = ct_counts.pivot(index = 'VClass', columns = 'trans_type', values = 'count')
type(ct_counts)
```

```python
sns.heatmap(ct_counts, annot = True, fmt= 'd'); #  use fmt = '.0f' if there are any cells with no counts, in order to account for NaNs
```

```python
sns.countplot(data = fuel_econ, x = 'VClass', hue='trans_type')
plt.xticks(rotation =15)

```

```python
ax = sns.countplot(data = fuel_econ, x = 'VClass', hue='trans_type')
ax.legend(loc = 2, ncol = 3, framealpha = 1, title = 'cat_var2');
plt.xticks(rotation =15);
```

```python
# Using bins and limiting 3 facset per row
bins = np.arange(12,58+2,2)
g = sns.FacetGrid(data=fuel_econ, col='VClass',col_wrap=3); # col_wrap limits number of facet to 3 in a row
g.map(plt.hist,'comb', bins=bins); 
```

- Use sharey= False to remove the mititation of having same Y axis. Each facet can have it's own Y axis
- Usage - Helps in rveiewing the change in distirbution of data over groups
- Issues- You lose out seeing the overall frequency of each group. One needs to pay attention to understand that the graphs are scaled fifferently and cannot be compared for frequency 

```python
# Set sharey=False
bins = np.arange(12,58+2,2)
g = sns.FacetGrid(data=fuel_econ, col='VClass',col_wrap=3, sharey=False); # col_wrap limits number of facet to 3 in a row
g.map(plt.hist,'comb', bins=bins); 
g.set_titles('{col_name}')
```

## Adapting Univariate plots to display bivariate relationsship
- Faceting
- Bar chart with mean outcome instead of the values (classifying into 0 and 1)
- Adaptive histogram

### Adaptive Bar Chart- 1 quantitative varaible against a qualitiative variable

```python
plt.figure(figsize=[15,10])

plt.subplot(3,3,1)

base_color = sns.color_palette()[0]
sns.barplot(data=fuel_econ, x='VClass', y='comb',color=base_color, errwidth=0); #errwidth=0 added to remove the uncertaiity whiskers
plt.xticks(rotation=15);

plt.subplot(3,3,2)

base_color = sns.color_palette()[0]
sns.barplot(data=fuel_econ, x='VClass', y='comb',color=base_color, ci='sd'); # ci ='sd' sets the wiskers to denote std deviation
plt.xticks(rotation=15);

plt.subplot(3,3,3)

# if having 0 lower bound is not important we can use poinplot
 
sns.pointplot(data=fuel_econ, x='VClass', y='comb', ci='sd'); # ci ='sd' sets the wiskers to denote std deviation
plt.xticks(rotation=15);

plt.subplot(3,3,4)

# if the plot was for nominal data instead of car classes which is ordinal (and sorted here), we should then remove the line
sns.pointplot(data=fuel_econ, x='VClass', y='comb', ci='sd', linestyles=''); # 
plt.xticks(rotation=15);
```

# Line Plots

- Summarixation for quant with Quant variables

```python
# set bin edges, compute centers
bin_edges = np.arange(0.6, 7+0.2,0.2)
bin_centers = bin_edges[:-1] + 0.1

displ_binned =pd.cut(fuel_econ['displ'],bin_edges, include_lowest=True)
comb_mean = fuel_econ['comb'].groupby(displ_binned).mean()
comb_std = fuel_econ['comb'].groupby(displ_binned).std()
```

```python
plt.errorbar(data=fuel_econ, x= bin_centers, y=comb_mean, yerr=comb_std);
plt.xticks(rotation=15);
plt.ylabel('Aveg Combined Fuel Eff. (mpg))');
```
